```python
import numpy as np   
wh=np.random.random((4,3))
bh=np.random.random((3,))
wout=np.random.random((3,))
bout=np.random.random((1,))

bout[0]=round(bout[0],2)


for i in range(4):
    for j in range(3):
        wh[i,j]=round(wh[i,j],2)

for i in range(3):
    bh[i]=round(bh[i],2)
    wout[i]=round(wout[i],2)

print("wh",wh)
print("bh",bh)
print("wout",wout)
print("bout",bout)
```
